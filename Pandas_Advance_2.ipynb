```python
# Q6. What do you understand about the windows function in pandas and list the types of windows functions?

'''
In Pandas, window functions (also known as rolling or moving functions) are used to perform calculations over a specified "window" or range of values in a DataFrame or Series. These functions are particularly useful for time-series analysis and other scenarios where you want to compute aggregated values over a moving set of data points. Window functions allow you to perform operations like moving averages, rolling sums, and more.

Window functions work by applying a function to a subset of data defined by a sliding window. The size of the window can be specified using a fixed number of periods or a time-based offset.

Here are some common types of window functions available in Pandas:

1. **Rolling Functions:**
   These functions operate over a rolling window of a fixed number of periods. They are particularly useful for time-series data.
   - `rolling().mean()`: Calculate the rolling mean.
   - `rolling().sum()`: Calculate the rolling sum.
   - `rolling().min()`: Calculate the rolling minimum.
   - `rolling().max()`: Calculate the rolling maximum.

2. **Expanding Functions:**
   These functions expand the window size over time. Each calculation includes all data up to the current position.
   - `expanding().mean()`: Calculate the expanding mean.
   - `expanding().sum()`: Calculate the expanding sum.
   - `expanding().min()`: Calculate the expanding minimum.
   - `expanding().max()`: Calculate the expanding maximum.

3. **Exponential Moving Average (EMA):**
   This is a weighted moving average where more recent data points are given higher weights.
   - `ewm().mean()`: Calculate the exponential moving average.

4. **Window Aggregations:**
   These functions perform custom aggregation operations on a window.
   - `rolling().apply(func)`: Apply a custom function to the rolling window.
   - `expanding().apply(func)`: Apply a custom function to the expanding window.

5. **Ranking Functions:**
   These functions compute the ranking of elements within a window.
   - `rolling().rank()`: Compute the rank of elements in a rolling window.
   - `expanding().rank()`: Compute the rank of elements in an expanding window.

6. **Shift and Lag Functions:**
   These functions allow you to access data from previous or future periods.
   - `shift()`: Shift the values of a Series by a specified number of periods.
   - `diff()`: Compute the difference between consecutive values.
   - `pct_change()`: Compute the percentage change between consecutive values.

These are just some of the window functions available in Pandas.
 Window functions provide powerful tools for analyzing and processing time-series and sequential data in a flexible and efficient manner.'''
```

```python
# Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days,
# hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

import pandas as pd

# Input dates in YYYY-MM-DD format
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Convert input strings to Pandas datetime objects
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

# Calculate the time difference using Timedelta
time_difference = date2 - date1

# Extract days, hours, and minutes from the time difference
days = time_difference.days
hours = time_difference.seconds // 3600
minutes = (time_difference.seconds // 60) % 60



# Display the result
print("Time Difference:")
print(f"Days: {days}")
print(f"Hours: {hours}")
print(f"Minutes: {minutes}")
```

```python
# Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
# column to a categorical data type. The program should prompt the user to enter the file path, column
# name, and category order, and then display the sorted data.

import pandas as pd

# Input file path, column name, and category order
file_path = input("Enter the CSV file path: ")
column_name = input("Enter the column name to convert to categorical: ")
category_order = input("Enter the category order (comma-separated): ").split(',')

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path)

# Convert the specified column to categorical
data[column_name] = pd.Categorical(data[column_name], categories=category_order, ordered=True)

# Sort the DataFrame based on the specified column
sorted_data = data.sort_values(by=[column_name])

# Display the sorted data
print("\nSorted Data:")
print(sorted_data)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Input file path
file_path = input("Enter the CSV file path: ")

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path)

# Convert the 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Pivot the data to create a DataFrame suitable for stacked bar chart
pivot_data = data.pivot(index='Date', columns='High', values='Low')

# Plot the stacked bar chart
pivot_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Sales by Product Category Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(title='Product Category')

# Show the chart
plt.show()
```
